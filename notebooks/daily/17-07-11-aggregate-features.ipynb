```python
def attributes(fp):
    t = fp.split('/')[-2]
    fp = fp.split('/')[-1]
    fp = fp.split('.')[0]
    try:
        m, ID, ps, l, ca = fp.split('/')[-1].split('_')
        ca = ca.split('.')[0].split('-')[1]
        l = l[1:]
        return t, m, ID, ps, l, ca
    except ValueError:
        m, ID, ps, l = fp.split('/')[-1].split('_')
        l = l[1:]
        return t, m, ID, ps, l
    
    
    
```

We would like two h5py files. One file containing all the features, and the other file containing the aggregated features.

**Features h5py file**

When accessing the features, there are 6 things to specify:
* Tissue
* ID
* Layer (of the network)
* Patch-size
* Model choice
* Convolutional aggregation (if layer != -1)

We want this is include all of the features that are in data/features. This might change and update. Therefore, we add each feature into an h5py file with these pieces of information as the keys.

**Aggregated Features h5py file**

* Tissue
* Patch-size
* Aggregation method
* Layer (of the network)

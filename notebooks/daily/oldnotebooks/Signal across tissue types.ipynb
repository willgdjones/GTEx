# Description

Here are the first five component associations for three different tissue types: Liver, Lung and Ovary. You'll notice that there seems to be signal for each of these tissues. Another interesting point is that there seems to be individual RNA transcripts that are associated to multiple components in a given tissue. For example, transcript ENSG00000135111.10 is one of the most associated transcripts in 3 of the first 5 components in the liver representations. Similarly, transcript ENSG000000204176.9 is one of the most associated transcripts in 2 of the first 5 components in the lung representations. This is probably because there is a lot of correlations among representations. It would be interesting to know what transcript are most associated across all components. Despite the follow up questions to be address, these figures do confirm that these associations are seen across tissue types.

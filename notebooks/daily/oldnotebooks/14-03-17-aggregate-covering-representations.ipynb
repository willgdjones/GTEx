```python
import os
import pickle
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

def get_donor_IDs(IDlist):
    return [str(x).split('-')[1] for x in IDlist]

IDs = os.listdir('../data/processed/assembled_representations/')
lung_expression_filepath = '/nfs/research2/stegle/stegle_secure/GTEx/download/49139/PhenoGenotypeFiles/RootStudyConsentSet_phs000424.GTEx.v6.p1.c1.GRU/ExpressionFiles/phe000006.v2.GTEx_RNAseq.expression-data-matrixfmt.c1/parse_data/44_tissues/GTEx_Data_20150112_RNAseq_RNASeQCv1.1.8_gene_rpkm_{}_normalised_without_inverse_gene_expression.txt'.format("Lung")

ID_representations = {}

for (j, ID) in enumerate(IDs):
    print (ID, len(IDs) - j)
    reps = pickle.load(open('../data/processed/assembled_representations/{}'.format(ID), 'rb'))
    ID_representations[ID] = reps
# pickle.dump(ID_representations,open('../data/processed/dictionaries/ID_reps','wb'))
# ID_representations = pickle.read(open('../data/processed/dictionaries/ID_reps','rb'))
    
```

```python
with open(lung_expression_filepath, 'r') as f:
    expression_table = np.array([x.split('\t') for x in f.read().splitlines()])
    expression_matrix = expression_table[1:,1:].astype(np.float32)

individual_lung_expression_donor_IDs = [x.split('-')[1] for x in expression_table[0,:][1:]]
print (len(individual_lung_expression_donor_IDs)) #278
    
```

```python
X = []
Y = []
for (gID, reps) in ID_representations.items():
    ID = str(gID).split('-')[1]
    print (ID)
    try:
#         print (ID)
        ID_idx = individual_lung_expression_donor_IDs.index(ID)
        X_row = expression_matrix[:,ID_idx]
        Y_row = np.median(ID_representations[gID], axis=0)
        X.append(X_row)
        Y.append(Y_row)
    except ValueError as e:
#         print(e)
        continue
        
X = np.vstack(X)
Y = np.vstack(Y)
pickle.dump([X,Y], open('../data/processed/associations/covering_data_median.py','wb'))
```
